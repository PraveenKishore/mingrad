# Mingrad visualizations

```python
import random
from mingrad.scalar import Scalar
from mingrad.nn import Neuron, MLP
from mingrad.visualize import draw_graph
```

### Visualizing basic operations

```python
a = Scalar(2)
b = Scalar(-6.5)
c = a*5 / 2
d = c + (b**6).relu()

d.backward()

draw_graph(d)
```

### Visualizing 2 neurons

```python
random.seed(42)

n = Neuron(2)
x = [Scalar(1.0), Scalar(-2.0)]
y = n(x)
y.backward()

dot = draw_graph(y)
dot
```
